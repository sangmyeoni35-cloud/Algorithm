```python
# 리스트에 저장된 숫자 중 가장 큰 숫자 출력 (최대값)

data_list = [23, 45, 21, 14, 3, 67]

max = data_list[0]
max_data = 0
for index in range(1, 6):
    if data_list[index] > max:
        max = data_list[index]
        max_index = index

print(max, max_index)
```

```python
# 리스트에 저장된 숫자 중 가장 작은 숫자 출력(최소값)

data_list = [23, 45, 21, 14, 3, 67]
target = 20
count = 0

for data in data_list:
    if data > target:
        count += 1

print(count)

```

```python
# 리스트에 저장된 모든 값 더하여 출력(합)

data_list = [23, 45, 21, 14, 3, 67]
total = 0

for index in range(6):
    total += data_list[index]

print(total)
```

```python
# 리스트에 저장된 숫자를 역순으로 저장하여 출력
data_list = [23, 45, 21, 14, 3, 67]

length = len(data_list)
index = 1

for data in data_list:
    temp = data_list[index - 1]
    data_list[index - 1] = data_list[length - index]
    data_list[length - index] = temp

for i in range(5):
    print(data_list[i])
```
