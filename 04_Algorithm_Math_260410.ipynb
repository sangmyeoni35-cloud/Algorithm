```python
def is_prime_number(n):
    if n < 2:
        return False
    
    for i in range(2, (n // 2) + 1):
        if n % i == 0:
            return False
            
    return True

num = int(input('숫자입력: '))
if is_prime_number(num):
          print(f"{num}은 소수입니다.")
else:
    print(f"{num}은 소수가 아닙니다.")
```

```python
### BOJ 11728
def merge_sorted_lists(A_list, B_list):
    # 각 리스트의 포인터(위치) 초기화
    pA = 0
    pB = 0
    N = len(A_list)
    M = len(B_list)
    
    # 결과가 담길 빈 리스트
    merged_list = []

    # --- 이미지 상단 로직: 두 값을 비교하며 작은 값부터 넣기 ---
    while pA < N and pB < M:
        if A_list[pA] < B_list[pB]:
            merged_list.append(A_list[pA])
            pA += 1
        else:
            merged_list.append(B_list[pB])
            pB += 1

    # --- 이미지 하단 로직: 한 리스트가 끝나면 남은 요소들 모두 넣기 ---
    # A_list에 남은 데이터가 있다면
    while pA < N:
        merged_list.append(A_list[pA])
        pA += 1
        
    # B_list에 남은 데이터가 있다면
    while pB < M:
        merged_list.append(B_list[pB])
        pB += 1

    return merged_list
```

```python
### BOJ 2003번
def count_subsequence_sum(num_list, M):
    N = len(num_list)
    answer = 0
    start = end = 0
    
    # 리스트가 비어있을 경우를 대비한 예외 처리
    if N == 0:
        return 0
        
    current_sum = num_list[0]
    
    while end < N:
        # 1. 현재 합(sum)이 목표치(M)와 같다면 정답 카운트 증가
        if current_sum == M:
            answer += 1
            
        # 2. 현재 합이 M보다 작거나 같으면 범위를 오른쪽으로 확장 (end 증가)
        if current_sum <= M:
            end += 1
            if end < N:
                current_sum += num_list[end]
        
        # 3. 현재 합이 M보다 크면 범위를 왼쪽에서 축소 (start 증가)
        else:
            current_sum -= num_list[start]
            start += 1
            
    return answer
```
