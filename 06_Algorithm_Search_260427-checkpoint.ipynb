```python
# while loop를 활용한 binary search
def binary_search(A, key):
    low = 0  #탐색영역의 시작
    high = len(key) - 1 # 탐색 영역의 끝

    while low <= high:
        # 중앙값을 찾는다
        middle = (low + high) // 2
        
        # 중앙에 있는 값과 비교하여 같으면 해당 index를 return
        if A[middle] == key: return middle
        # 작으면 -> 왼쪽 영역 탐색 high 값 변경
        elif key < A[middle]:
            high = middle - 1
        # 크면 오른쪽 영역 탐색 low 값 변경
        else:
            low = middle + 1

    return -1 # 찾는 데이터가 없음
    
```
